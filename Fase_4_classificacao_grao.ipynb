```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar as features (X) e o alvo (y)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar o modelo Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Fazer previsões
y_pred = clf.predict(X_test)

# Avaliação das métricas
print("ACURÁCIA:", accuracy_score(y_test, y_pred))
print("PRECISÃO (média macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (média macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (média macro):", f1_score(y_test, y_pred, average='macro'))

# Relatório completo
print("\nRELATÓRIO DE CLASSIFICAÇÃO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap='Blues')
plt.title("Matriz de Confusão - Random Forest")
plt.show()
```

```python
#Treinar com Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar as features (X) e o alvo (y)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar modelo de Regressão Logística
logreg = LogisticRegression(max_iter=200, random_state=42)
logreg.fit(X_train, y_train)

# Fazer previsões
y_pred = logreg.predict(X_test)

# Avaliação das métricas
print("ACURÁCIA:", accuracy_score(y_test, y_pred))
print("PRECISÃO (média macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (média macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (média macro):", f1_score(y_test, y_pred, average='macro'))

# Relatório de classificação
print("\nRELATÓRIO DE CLASSIFICAÇÃO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=logreg.classes_)
disp.plot(cmap='Oranges')
plt.title("Matriz de Confusão - Logistic Regression")
plt.show()
```

```python
#Treinar com K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar X (features) e y (alvo)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar o modelo KNN (com k=5, valor padrão comum)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Prever no conjunto de teste
y_pred = knn.predict(X_test)

# Avaliação
print("ACURÁCIA:", accuracy_score(y_test, y_pred))
print("PRECISÃO (macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (macro):", f1_score(y_test, y_pred, average='macro'))

# Relatório completo
print("\nRELATÓRIO DE CLASSIFICAÇÃO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de Confusão
cm = confusion_matrix(y_test, y_pred, labels=knn.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(cmap='Blues')
plt.title("Matriz de Confusão - KNN")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Criar um dicionário para armazenar os resultados
resultados = {}

# Lista de modelos já treinados
modelos = {
    'Random Forest': clf,
    'Logistic Regression': logreg,
    'KNN': knn
}

# Prever e calcular métricas para cada modelo
for nome, modelo in modelos.items():
    y_pred = modelo.predict(X_test)
    resultados[nome] = {
        'Acurácia': accuracy_score(y_test, y_pred),
        'Precisão': precision_score(y_test, y_pred, average='macro'),
        'Recall': recall_score(y_test, y_pred, average='macro'),
        'F1-Score': f1_score(y_test, y_pred, average='macro')
    }

# Mostrar resultados como DataFrame
import pandas as pd
df_resultados = pd.DataFrame(resultados).T
print(df_resultados)
```

```python
#O Logistic regression mostrou uma boa precisão, mas segue abaixo como seria uma otimização com Grid Search

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Reutilizando X e y
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Separar dados
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

param_clf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
}

grid_clf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid=param_clf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_clf.fit(X_train, y_train)
print("Melhores parâmetros (Random Forest):", grid_clf.best_params_)

pipe_logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=1000))
])

param_logreg = {
    'logreg__C': [0.01, 0.1, 1, 10],
    'logreg__penalty': ['l2'],
    'logreg__solver': ['lbfgs'],
}

grid_logreg = GridSearchCV(
    pipe_logreg,
    param_grid=param_logreg,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_logreg.fit(X_train, y_train)
print("Melhores parâmetros (Logistic Regression):", grid_logreg.best_params_)


pipe_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

param_knn = {
    'knn__n_neighbors': [3, 5, 7, 9],
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2]  # p=1: Manhattan, p=2: Euclidean
}

grid_knn = GridSearchCV(
    pipe_knn,
    param_grid=param_knn,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_knn.fit(X_train, y_train)
print("Melhores parâmetros (KNN):", grid_knn.best_params_)

```

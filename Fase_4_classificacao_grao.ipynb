```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar as features (X) e o alvo (y)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar o modelo Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Fazer previs√µes
y_pred = clf.predict(X_test)

# Avalia√ß√£o das m√©tricas
print("ACUR√ÅCIA:", accuracy_score(y_test, y_pred))
print("PRECIS√ÉO (m√©dia macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (m√©dia macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (m√©dia macro):", f1_score(y_test, y_pred, average='macro'))

# Relat√≥rio completo
print("\nRELAT√ìRIO DE CLASSIFICA√á√ÉO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de confus√£o
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap='Blues')
plt.title("Matriz de Confus√£o - Random Forest")
plt.show()
```

```python
#Treinar com Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar as features (X) e o alvo (y)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar modelo de Regress√£o Log√≠stica
logreg = LogisticRegression(max_iter=200, random_state=42)
logreg.fit(X_train, y_train)

# Fazer previs√µes
y_pred = logreg.predict(X_test)

# Avalia√ß√£o das m√©tricas
print("ACUR√ÅCIA:", accuracy_score(y_test, y_pred))
print("PRECIS√ÉO (m√©dia macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (m√©dia macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (m√©dia macro):", f1_score(y_test, y_pred, average='macro'))

# Relat√≥rio de classifica√ß√£o
print("\nRELAT√ìRIO DE CLASSIFICA√á√ÉO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de confus√£o
cm = confusion_matrix(y_test, y_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=logreg.classes_)
disp.plot(cmap='Oranges')
plt.title("Matriz de Confus√£o - Logistic Regression")
plt.show()
```

```python
#Treinar com K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Separar X (features) e y (alvo)
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Dividir os dados em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar o modelo KNN (com k=5, valor padr√£o comum)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Prever no conjunto de teste
y_pred = knn.predict(X_test)

# Avalia√ß√£o
print("ACUR√ÅCIA:", accuracy_score(y_test, y_pred))
print("PRECIS√ÉO (macro):", precision_score(y_test, y_pred, average='macro'))
print("RECALL (macro):", recall_score(y_test, y_pred, average='macro'))
print("F1-SCORE (macro):", f1_score(y_test, y_pred, average='macro'))

# Relat√≥rio completo
print("\nRELAT√ìRIO DE CLASSIFICA√á√ÉO:\n")
print(classification_report(y_test, y_pred, target_names=['Kama', 'Rosa', 'Canadian']))

# Matriz de Confus√£o
cm = confusion_matrix(y_test, y_pred, labels=knn.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(cmap='Blues')
plt.title("Matriz de Confus√£o - KNN")
plt.show()
```

```python
#O Logistic regression mostrou uma boa precis√£o, mas segue abaixo como seria uma otimiza√ß√£o com Grid Search

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Reutilizando X e y
X = dados.drop(columns=['Variedade'])
y = dados['Variedade']

# Separar dados
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

param_clf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
}

grid_clf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid=param_clf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_clf.fit(X_train, y_train)
print("Melhores par√¢metros (Random Forest):", grid_clf.best_params_)

pipe_logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=1000))
])

param_logreg = {
    'logreg__C': [0.01, 0.1, 1, 10],
    'logreg__penalty': ['l2'],
    'logreg__solver': ['lbfgs'],
}

grid_logreg = GridSearchCV(
    pipe_logreg,
    param_grid=param_logreg,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_logreg.fit(X_train, y_train)
print("Melhores par√¢metros (Logistic Regression):", grid_logreg.best_params_)


pipe_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

param_knn = {
    'knn__n_neighbors': [3, 5, 7, 9],
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2]  # p=1: Manhattan, p=2: Euclidean
}

grid_knn = GridSearchCV(
    pipe_knn,
    param_grid=param_knn,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid_knn.fit(X_train, y_train)
print("Melhores par√¢metros (KNN):", grid_knn.best_params_)

```
